**c) Save the model**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state =1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.impute import SimpleImputer
```

```python
numeric_processor = Pipeline(
    steps = [("imputation_mean", SimpleImputer(missing_values = np.nan, strategy = "mean")),
              ("scaler", StandardScaler())]
)
```

```python
numeric_processor
```

```python
# categorical processing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor = Pipeline(
   steps = [("imputation_constant", SimpleImputer(fill_value = "missing", strategy = "constant")),
                                         ("onehot", OneHotEncoder(handle_unknown = "ignore"))]
)
```

```python
categorical_processor
```

**d)To apply dimensionality reduction technique- PCA on a dataset and compare and contrast the difference in performance of the algorithm with and without PCA**

```python
from sklearn.compose import ColumnTransformer
```

```python
# from sklearn.decomposition import PCA
preprocessor = ColumnTransformer(
    [("categorical", categorical_processor, ["type"]),
     ("numerical", numeric_processor, ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "density"])]
)
```

```python
df.info()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(preprocessor, LogisticRegression())
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python

#confusion matrix
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
import pickle
```

```python
# save the classification model as pickle file
model_pkl_file = "Telecom_classifier_model.pkl"
with open(model_pkl_file, 'wb') as file:
  pickle.dump(pipe, file)
```

```python
# load model from pickle file
with open(model_pkl_file, 'rb') as file:
  model = pickle.load(file)

# evaluate model
y_predict = model.predict(X_test)
```

```python
print(metrics.classification_report(y_test, y_predict))
```

```python
#LAMBDA FUNCTION
df['mainroad'] = df.apply(lambda x: 1 if x['mainroad'] == 'yes' else 0, axis = 1)
```

```python
df['basement'] = df.apply(lambda x: 1 if x['basement'] == 'yes' else 0, axis = 1)
```

```python
df['guestroom'] = df.apply(lambda x: 1 if x['guestroom'] == 'yes' else 0, axis = 1)
```

```python
#USING MAP
df['furnishingstatus'] = df['furnishingstatus'].map({'unfurnished':0, 'semi-furnished': 1, 'furnished': 2})
```

```python
df['hotwaterheating'] = df.apply(lambda x: 1 if x['hotwaterheating'] == 'yes' else 0, axis = 1)
df['airconditioning'] = df.apply(lambda x: 1 if x['airconditioning'] == 'yes' else 0, axis = 1)
df['prefarea'] = df.apply(lambda x: 1 if x['prefarea'] == 'yes' else 0, axis = 1)
```

```python
X_train['mainroad'] = pd.to_numeric(X_train['mainroad'], errors='coerce')
X_train['guestroom'] = pd.to_numeric(X_train['guestroom'], errors='coerce')
X_train['basement'] = pd.to_numeric(X_train['basement'], errors='coerce')
X_train['hotwaterheating'] = pd.to_numeric(X_train['hotwaterheating'], errors='coerce')
X_train['airconditioning'] = pd.to_numeric(X_train['airconditioning'], errors='coerce')
X_train['furnishingstatus'] = pd.to_numeric(X_train['furnishingstatus'], errors='coerce')
X_train['prefarea'] = pd.to_numeric(X_train['prefarea'], errors='coerce')

```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin = lin.fit(X_train, y_train)
y_pred = lin.predict(X_test)
```

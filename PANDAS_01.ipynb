```python
#1. Load the dataset in the jupyter notebook using pandas.
import pandas as pd
```

```python
df = pd.read_csv("/content/Housing.csv")
```

```python
#2. Print the following information:
#a) First five rows,
#b) Last ten rows,
#c) Name of all the columns,
#d) Shape of the dataset,
#e) Data types of all the columns














print("First five rows:")
print(df.head())


print("\nLast ten rows:")
print(df.tail(10))


print("\nColumn names:")
print(df.columns)

print("\nShape of the dataset:")
print(df.shape)


print("\nData types of columns:")
print(df.dtypes)
```

```python
#3. Create a new data frame of
#a) All categorical values
#b) All numerical values

categorical_df = df.select_dtypes(include=['object'])
numerical_df = df.select_dtypes(include=['int64', 'float64'])
print("Categorical dataframe:")
print(categorical_df.head())

print("\nNumerical dataframe:")
print(numerical_df.head())
```

```python
#4. Print the:
#a) No. of unique values of  “furnishingstatus” and their count
#b) No. of unique values of  “stories” and their count


furnishingstatus_counts = df['furnishingstatus'].value_counts()
print("Unique values and counts for 'furnishingstatus':")
print(furnishingstatus_counts)

stories_counts = df['stories'].value_counts()
print("\nUnique values and counts for 'stories':")
print(stories_counts)
```

```python
#5. Print the:
#a) Price of the houses which are furnished.
#b) Area and Price of houses having bathroom >= 3 and guestroom = yes
#c) Rows where the price of the houses is more than 100 times of their areas.
#d) Mean price of the houses having furnishingstatus = semi-furnished

# a)
furnished_prices = df[df['furnishingstatus'] == 'furnished']['price']
print("Price of furnished houses:")
print(furnished_prices)

# b)
filtered_houses = df[(df['bathrooms'] >= 3) & (df['guestroom'] == 'yes')]
filtered_area_price = filtered_houses[['area', 'price']]
print("\nArea and Price of houses with bathroom >= 3 and guestroom = yes:")
print(filtered_area_price)

# c)
price_area_ratio = df[df['price'] > 100 * df['area']]
print("\nRows where price is more than 100 times of area:")
print(price_area_ratio)

# d)
semi_furnished_mean_price = df[df['furnishingstatus'] == 'semi-furnished']['price'].mean()
print("\nMean price of houses with furnishingstatus = semi-furnished:", semi_furnished_mean_price)
```

```python
#6. Print the:
#a) Number of missing values in each column
#b) Code to replace the missing values in column = bedrooms with mean
#c) Code to replace the missing values in column = stories with the median
#d) Code to replace the missing values in column = furnishingstatus with appropriate central tendency.

# a)
missing_values_count = df.isnull().sum()
print("Number of missing values in each column:")
print(missing_values_count)

# b)
mean_bedrooms = df['bedrooms'].mean()
df['bedrooms'].fillna(mean_bedrooms, inplace=True)

# c)
median_stories = df['stories'].median()
df['stories'].fillna(median_stories, inplace=True)

# d)
mode_furnishingstatus = df['furnishingstatus'].mode()[0]
df['furnishingstatus'].fillna(mode_furnishingstatus, inplace=True)

print("\nNumber of missing values after replacement:")
print(df.isnull().sum())
```

```python
#7. Due to a sudden hike in land prices, all the land prices hiked 10% of their previous price. Reflect this change in the original dataset.

df.loc[df['mainroad'] == 'yes', 'price'] = df[df['mainroad'] == 'yes']['price'] * 1.1

print("Updated dataset with land prices increased by 10%:")
print(df.head())

updated_dataset_path = 'Updated_Housing.csv'
df.to_csv(updated_dataset_path, index=False)
```

```python
#8. Save this updated data frame in your system as “updated_housing.xlsx”

updated_excel_path = 'updated_housing.xlsx'
df.to_excel(updated_excel_path, index=False)

print("Updated dataframe saved as 'updated_housing.xlsx'")
```
